```python
item_weight = sales_df.pivot_table(values='Item_Weight', index='Item_Identifier')

is_null = sales_df['Item_Weight'].isnull()

temp = pd.DataFrame(sales_df) 
temp[is_null,'Item_Weight'] = sales_df.loc[is_null,'Item_Identifier'].apply(lambda x:  item_weight.at[x,'Item_Weight'])
temp
```
